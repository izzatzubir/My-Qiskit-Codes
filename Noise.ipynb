```python
from qiskit import * #from qiskit repo, import everything
%matplotlib inline #used to printout diagrams using mpl
```

```python
nqubits = 3 #set variable nqubits to 3
circuit = QuantumCircuit(nqubits, nqubits) #set variable circuit to become a Quantum circuit with (qubits, bit)
circuit.h(0) #Hadamard gate to first qubit
circuit.cx(0,1) #C-NOT gate to first qubit as control, and second qubit as target
circuit.cx(1,2) #C-NOT gate to second qubit as control, and third qubit as target
circuit.measure([0,1,2],[0,1,2]) #measure the qubits, to the respective classical bits
```

```python
circuit.draw(output ='mpl', initial_state= True) #Draw the circuit, and display the initial state of the qubits
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
print(statevector)
%matplotlib inline
circuit.draw(output = 'mpl')
```
